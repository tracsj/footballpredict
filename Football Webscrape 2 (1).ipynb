### Background
Predicting outcomes of Football games is both an interesting task, given that football is a game comprised of many more complex play than other sports and where 'unexpected' results occur frequently, and also one with considerable potential value given the betting markets that exist around the sport. As a fan of the sport I am also interested to see if any unexpected features present themselves as important factors for determining the outcome of a football match.

>Soccer isn't baseball, in other words, and even leading researchers in soccer numerology doubt they'll discover a universally applicable set of metrics—there are simply too many differences between teams and cultures, and too much chaotic complexity within the game itself.”

### Hypothesis
My initial hypothesis is that team features related to prior performance will act as decent predictors of match outcomes. My further hypothesis is that if we use player level features relevant to each match e.g. take into account the players that will actually perform in each game, we will yield more accurate predictions. This hypothesis is based on the insight that player level performance features are more representative of the team playing in a given match than team level features representing all players that have played across the season.

```python
class ResultsTools(object):
    """
    Performs various functions on input x to extract key match info from complex string
    
    Attributes:
      x = string item containing match info coming from the team performance dictionary 
    
    Returns extracted results relevant to match encoded in string x
    """
    
    def __init__(self):
        self.x = object
            
    def result(self,x):
    #determine if match was won or lost by club_name
        if x.find('win') != -1:
            return'win'
        elif x.find('lose') != -1:
            return 'lose'
        else:
            return 'draw'
        
    def game_date(self,x):
        a = re.findall("(\d*-\d*-\d*).*",x)
        if a[0][0] == str(0):
             a = a[0][1:].replace('-0','/').replace('-','/')
        else:
            a = a[0].replace('-0','/').replace('-','/')
        return a
    
    def game_opponent(self,x):
        a = re.findall("\>(\w*.*\w*)\<",x)
        return a[0]
    
    def game_homeaway(self,x):
        a = re.findall("\</a>(\d*)",x)
        if len(a[0]) > 0:
            return 'home'
        else:
            return 'away'
```
